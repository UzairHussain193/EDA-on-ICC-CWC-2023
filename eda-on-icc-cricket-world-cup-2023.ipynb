## <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Pacifico; background-color:RoyalBlue; overflow:hidden"><b>4.4.1 Overall Batsman Performance</b></div>

<div style="padding: 20px; border-color: #c77220; border-radius: 10px; box-shadow: 0 2px 4px 0 rgba(0, 0, 0, 0.1); border: 2px solid RoyalBlue;">
    <h3>1st Function which is (OverallBatsmanStats) will calculate the overall tournament stats:</h3>
      <ul>
        <li>Player: Name of the player</li>
        <li>Matches: Total number of matches played</li>
        <li>Score: Total number of runs scored</li>
        <li>Avg: Average</li>
        <li>SR: Strike rate</li>
        <li>Balls Faced: Total number of balls played</li>
        <li>HS: Highest score</li>
        <li>50s: Total number of 50's scored</li>
        <li>100s: Total number of 100's scored</li>
        <li>Dismissed: In how many matches the player got out</li>
      </ul>
</div>

```python
def OverallBatsmanStats(p_name):    
    player = p_name

    # creating a list of players who played ICC CWC_2023
    playerList = df['striker'].unique().tolist()
    
    #Checking if the player name is in list of players who played ICC CWC_2023
    if player in playerList:
        batsman = df[df['striker'] == player]
        
        # Find number of matches played by a player
        matches = batsman['match_id'].nunique()
        
        # Find number of runs scored by a player
        score = batsman['runs_off_bat'].sum()
        
        # Find number of 50's scored by a player
        runs = batsman.groupby('match_id')['runs_off_bat'].sum().values.tolist()

        fiftees = 0
        hundreds = 0

        for i in runs:
            if i >= 50 and i <100:
                fiftees += 1
                
            elif i >= 100:
                hundreds += 1
                        
        # Find the highest runs scored by a player
        highestScore = batsman.groupby('match_id')['runs_off_bat'].sum().max()
        
        # Find the number of times a batsman has been dismissed
        dismissed = batsman.player_dismissed.count()
        
        # Find number of Balls Faced by a player
        ballsFaced = batsman['runs_off_bat'].count()
        
        # Find SR of Batsman
        sr = (score/ballsFaced)*100
        
        # Find Average of Player
        avg = score/dismissed
        
        data = {
            'PlayeName': [player],
            'Matches' : [matches],
            'Score' : [score],
            'Avg' : [round(avg,3)],
            'SR' : [round(sr,3)],
            'BallsFaced': [ballsFaced],
            'HS' : [highestScore],
            '50s' : [fiftees],
            '100s' : [hundreds],
            'Dismissed': [dismissed]
        }
        
        Stats_df = pd.DataFrame(data)
                
        return Stats_df
    
    else:
        return 'Invalid Player Name, Please Re-Enter the name of Player or Check the spelling of the name'
```

<div style="padding: 20px; border-color: #c77220; border-radius: 10px; box-shadow: 0 2px 4px 0 rgba(0, 0, 0, 0.1); border: 2px solid RoyalBlue;">
    <h3>2nd Function which is (AgainstEachTeamBatsmanAnalysis) will calculate only for a single match:</h3>
      <ul>
        <li>Scores: Total number of runs scored in the match</li>
        <li>Balls: Total number of balls faced in the match</li>
        <li>Opponent: Which was the opponent team that played</li>
        <li>Venue: Where the match was played</li>
      </ul>
</div>


```python
def AgainstEachTeamBatsmanAnalysis(p_name):
    player = p_name
    
    # creating a list of players who played ICC CWC_2023
    playerList = df['striker'].unique().tolist()
    
    #Checking if the player name is in list of players who played ICC CWC_2023
    if player in playerList:
        
        batsman = df[df['striker'] == player]
        
        matches_played = sorted(batsman['match_id'].unique().tolist())
        
        matches_played

        # Create empty lists to store the scores, balls faced, opponent team, and venue for each match    
        scores = []
        balls = []
        opponent = []
        venue = []

        for i in matches_played:
            # Calculate the total runs scored by the player in the current match and append it to the scores list
            scores.append(batsman[batsman['match_id'] == i]['runs_off_bat'].sum())
            # Count the total number of balls faced by the player in the current match and append it to the balls list
            balls.append(batsman[batsman['match_id'] == i]['runs_off_bat'].count())
            # Get the opponent team for the current match and append it to the opponent list
            opponent.append(batsman[batsman['match_id'] == i]['bowling_team'].unique()[0])
            # Extract the venue of the current match, split it to get the city name, and append it to the venue list
            venue.append(batsman[batsman['match_id'] == i]['venue'].unique()[0].split(',')[-1])
        
        teamWise = {
            'Scores' : scores,
            'Balls' : balls,
            'Opponent' : opponent,
            'Venue' : venue
        }
        
        
        TeamWiseStats = pd.DataFrame(teamWise)
        
        return TeamWiseStats
    
    else:
        return 'Invalid Player Name, Please Re-Enter the name of Player or Check the spelling of the name'
```

### 2nd Function which is (OverallBowlerStats) will calculate the overall tournament stats for bowlers:
- BowlerName: Name of the bowler
- Matches: Total number of matches played
- Wickets: Total number of wickets taken
- RunsConceded: Total number of runs conceded
- BallsBowled: Total number of balls bowled
- Economy: Runs conceded per over
- Average: Runs conceded per wicket
- StrikeRate: Balls bowled per wicket


```python
def OverallBowlerStats(bowler_name):
    bowler = bowler_name

    # creating a list of bowlers who played ICC CWC_2023
    bowlerList = df['bowler'].unique().tolist()
    
    # Checking if the bowler name is in the list of bowlers who played ICC CWC_2023
    if bowler in bowlerList:
        bowler_data = df[df['bowler'] == bowler]
        
        # Find number of matches played by the bowler
        matches = bowler_data['match_id'].nunique()
        
        # Find total wickets taken by the bowler
        wickets = bowler_data['player_dismissed'].count()
        
        # Find total runs conceded by the bowler
        runs_conceded = bowler_data['runs_off_bat'].sum() + bowler_data['extras'].sum()
        
        # Find total balls bowled by the bowler
        balls_bowled = bowler_data['ball'].count()
        
        # Calculate economy rate
        economy = (runs_conceded / (balls_bowled / 6)) if balls_bowled > 0 else 0
        
        # Calculate average (runs conceded per wicket)
        avg = (runs_conceded / wickets) if wickets > 0 else 0
        
        # Calculate strike rate (balls per wicket)
        strike_rate = (balls_bowled / wickets) if wickets > 0 else 0
        
        data = {
            'BowlerName': [bowler],
            'Matches': [matches],
            'Wickets': [wickets],
            'RunsConceded': [runs_conceded],
            'BallsBowled': [balls_bowled],
            'Economy': [round(economy, 2)],
            'Average': [round(avg, 2)],
            'StrikeRate': [round(strike_rate, 2)]
        }
        
        Stats_df = pd.DataFrame(data)
        
        return Stats_df
    
    else:
        return 'Invalid Bowler Name, Please Re-Enter the name of Bowler or Check the spelling of the name'
```

### 2nd Function which is (AgainstEachTeamBowlerAnalysis) will calculate only for a single match:
- Wickets: Total number of wickets taken against each team
- Runs Conceded: Total runs conceded against each team
- Balls Bowled: Total No of Balls Bowled in single match
- Opponent: Which was the opponent team that played
- Venue: Where the match was played


```python
def AgainstEachTeamBowlerAnalysis(bowler_name):
    bowler = bowler_name
    
    # creating a list of bowlers who played ICC CWC_2023
    bowlerList = df['bowler'].unique().tolist()
    
    # Checking if the bowler name is in the list of bowlers who played ICC CWC_2023
    if bowler in bowlerList:
        
        bowler_data = df[df['bowler'] == bowler]
        
        matches_played = sorted(bowler_data['match_id'].unique().tolist())
        
        # Create empty lists to store the wickets, runs conceded, balls bowled, opponent team, and venue for each match    
        wickets = []
        runs_conceded = []
        balls_bowled = []
        opponent = []
        venue = []

        for i in matches_played:
            # Calculate the total wickets taken by the bowler in the current match and append it to the wickets list
            wickets.append(bowler_data[bowler_data['match_id'] == i]['player_dismissed'].count())
            # Calculate the total runs conceded by the bowler in the current match and append it to the runs_conceded list
            runs_conceded.append(bowler_data[bowler_data['match_id'] == i]['runs_off_bat'].sum() + bowler_data[bowler_data['match_id'] == i]['extras'].sum())
            # Count the total number of balls bowled by the bowler in the current match and append it to the balls_bowled list
            balls_bowled.append(bowler_data[bowler_data['match_id'] == i]['ball'].count())
            # Get the opponent team for the current match and append it to the opponent list
            opponent.append(bowler_data[bowler_data['match_id'] == i]['batting_team'].unique()[0])
            # Extract the venue of the current match, split it to get the city name, and append it to the venue list
            venue.append(bowler_data[bowler_data['match_id'] == i]['venue'].unique()[0].split(',')[-1])
        
        teamWise = {
            'Wickets': wickets,
            'RunsConceded': runs_conceded,
            'BallsBowled': balls_bowled,
            'Opponent': opponent,
            'Venue': venue
        }
        
        TeamWiseStats = pd.DataFrame(teamWise)
        
        return TeamWiseStats
    
    else:
        return 'Invalid Bowler Name, Please Re-Enter the name of Bowler or Check the spelling of the name'
```
